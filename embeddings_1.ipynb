```python
#imports

from sentence_transformers import SentenceTransformer
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#text data
questions = [
  "How do I learn python?",
  "What is the best way to learn machine learning?",
  "How do i bake a chocolate cake?",
  "Where can I find a good python tutorial?",
  "How do I cook pasta?",
  "Tips for learning programming with python?",
  "What ingredients do I need for a pasta?",  
  "What are some beginner resources for learning python?"
]
```

```python
#gen sentence embeddings
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(questions)
```

```python
#reduce dimensions with t-SNE
tsne = TSNE(n_components=2, perplexity=5, random_state=42)  
embeddings_2d = tsne.fit_transform(embeddings)
```

```python
#plot the results
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")

for i,(x, y) in enumerate(embeddings_2d):
    plt.scatter(x, y, color='steelblue')
    plt.text(x + 0.5, y, questions[i], fontsize=9)
plt.title('t-SNE Visualization of Sentence Embeddings')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.tight_layout()
plt.show()


#save the plot
```
